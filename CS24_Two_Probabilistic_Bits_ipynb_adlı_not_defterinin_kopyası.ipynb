Suppose that we have two probabilistic bits, and our probabilistic states respectively are

$ \myvector{0.2 \\ 0.8} \mbox{ and } \myvector{0.6 \\ 0.4 }. $

If we combine both bits as a single system, then what is the state of the combined system?

In total, we have four different states. We can name them as follows:
<ul>
    <li>00: both bits are in states 0</li>
    <li>01: the first bit is in state 0 and the second bit is in state 1</li>
    <li>10: the first bit is in state 1 and the second bit is in state 0</li>
    <li>11: both bits are in states 1</li>
</ul>

<h3> Task 1 </h3>

<b>Discussion and analysis:</b>

What are the probabilities of being in states $ 00 $, $ 01 $, $ 10 $, and $11$?

How can we represent these probabilities as a column vector?

<h3> Representation for states 0 and 1</h3>

The vector representation of state 0 is $ \myvector{1 \\ 0} $. Similarly, the vector representation of state 1 is $ \myvector{0 \\ 1} $.

We use $ \pstate{0} $ to represent $ \myvector{1 \\ 0} $ and $ \pstate{1} $ to represent $ \myvector{0 \\ 1} $.

Then, the probabilistic state $ \myvector{0.2 \\ 0.8} $ is also represented as $ 0.2 \pstate{0} + 0.8 \pstate{1} $.

Similarly, the probabilistic state $ \myvector{0.6 \\ 0.4} $ is also represented as $ 0.6 \pstate{0} + 0.4 \pstate{1} $.

Başlangıç Durumları (Verilenler)$$Bit_1 = \begin{pmatrix} 0.2 \\ 0.8 \end{pmatrix} \quad (0 \text{ gelme olasiligi: } 0.2, \ 1 \text{ gelme olasiligi: } 0.8)$$$$Bit_2 = \begin{pmatrix} 0.6 \\ 0.4 \end{pmatrix} \quad (0 \text{ gelme olasiligi: } 0.6, \ 1 \text{ gelme olasiligi: } 0.4)$$Görev 1: Olasılıkların Hesaplanması (Tensör Çarpımı)İki bağımsız olayın birlikte gerçekleşme olasılığı çarpılarak bulunur.Durum 00 (Birinci 0, İkinci 0):$$Pr(00) = Pr(Bit_1=0) \times Pr(Bit_2=0)$$$$Pr(00) = 0.2 \times 0.6 = \mathbf{0.12}$$Durum 01 (Birinci 0, İkinci 1):$$Pr(01) = Pr(Bit_1=0) \times Pr(Bit_2=1)$$$$Pr(01) = 0.2 \times 0.4 = \mathbf{0.08}$$Durum 10 (Birinci 1, İkinci 0):$$Pr(10) = Pr(Bit_1=1) \times Pr(Bit_2=0)$$$$Pr(10) = 0.8 \times 0.6 = \mathbf{0.48}$$Durum 11 (Birinci 1, İkinci 1):$$Pr(11) = Pr(Bit_1=1) \times Pr(Bit_2=1)$$$$Pr(11) = 0.8 \times 0.4 = \mathbf{0.32}$$Görev 2: Sütun Vektörü Olarak GösterimBulduğumuz sonuçları sırasıyla (00, 01, 10, 11) alt alta yazarız:$$v_{kompozit} = \begin{pmatrix} 0.12 \\ 0.08 \\ 0.48 \\ 0.32 \end{pmatrix}$$(Normalizasyon Kontrolü: $0.12 + 0.08 + 0.48 + 0.32 = 1.0$)Görev 3: Vektörü Temel Durumlarla (Basis States) İfade EtmeBirinci bitin durumunu $\begin{pmatrix} 0.2 \\ 0.8 \end{pmatrix}$ parçalayarak yazımı:$$\begin{pmatrix} 0.2 \\ 0.8 \end{pmatrix} = \begin{pmatrix} 0.2 \\ 0 \end{pmatrix} + \begin{pmatrix} 0 \\ 0.8 \end{pmatrix}$$Skaler çarpanları dışarı alırız:$$= 0.2 \begin{pmatrix} 1 \\ 0 \end{pmatrix} + 0.8 \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$Soru bizden $\begin{pmatrix} 1 \\ 0 \end{pmatrix}$ yerine $[0]$, $\begin{pmatrix} 0 \\ 1 \end{pmatrix}$ yerine $[1]$ yazmamızı istiyor:$$= \mathbf{0.2[0] + 0.8[1]}$$


<h3> Composite systems </h3>

When two systems are composed, then their states are tensored to calculate the state of composite system.

The probabilistic state of the first bit is $  \myvector{0.2 \\ 0.8} = 0.2 \pstate{0} + 0.8 \pstate{1} $.

The probabilistic state of the second bit is $  \myvector{0.6 \\ 0.4} = 0.6 \pstate{0} + 0.4 \pstate{1} $.

Then, the probabilistic state of the composite system is $ \big( 0.2 \pstate{0} + 0.8 \pstate{1} \big) \otimes  \big( 0.6 \pstate{0} + 0.4 \pstate{1} \big) $.

### Task 2: Dağılma Özelliği (Adım Adım FOIL Yöntemi)

Elimizdeki ifade matematikteki $(A + B) \times (C + D)$ çarpımının aynısıdır. Birinci parantezdeki her elemanı, ikinci parantezdeki her elemanla tek tek çarpıp birleştireceğiz.

İfademiz:
$$(\mathbf{0.2[0]} + \mathbf{0.8[1]}) \otimes (\mathbf{0.6[0]} + \mathbf{0.4[1]})$$

**1. Adım: Birinci Terimi Dağıtmak ($0.2[0]$)**
Birinci parantezin ilk elemanını ($0.2[0]$) alıp, karşıdaki iki elemanla sırasıyla çarpıyoruz:
* İlk terim $\times$ İlk terim:
    $$0.2[0] \otimes 0.6[0] \implies \text{Say\i lar\i\ \c{c}arp: } (0.2 \times 0.6 = 0.12) \implies \text{Durumlar\i\ birle\c{s}tir: } [0 \text{ yan\i na } 0] \implies \mathbf{0.12[00]}$$
* İlk terim $\times$ İkinci terim:
    $$0.2[0] \otimes 0.4[1] \implies \text{Say\i lar\i\ \c{c}arp: } (0.2 \times 0.4 = 0.08) \implies \text{Durumlar\i\ birle\c{s}tir: } [0 \text{ yan\i na } 1] \implies \mathbf{0.08[01]}$$

**2. Adım: İkinci Terimi Dağıtmak ($0.8[1]$)**
Şimdi birinci parantezin ikinci elemanını ($0.8[1]$) alıp, karşıdaki iki elemanla çarpıyoruz:
* İkinci terim $\times$ İlk terim:
    $$0.8[1] \otimes 0.6[0] \implies \text{Say\i lar\i\ \c{c}arp: } (0.8 \times 0.6 = 0.48) \implies \text{Durumlar\i\ birle\c{s}tir: } [1 \text{ yan\i na } 0] \implies \mathbf{0.48[10]}$$
* İkinci terim $\times$ İkinci terim:
    $$0.8[1] \otimes 0.4[1] \implies \text{Say\i lar\i\ \c{c}arp: } (0.8 \times 0.4 = 0.32) \implies \text{Durumlar\i\ birle\c{s}tir: } [1 \text{ yan\i na } 1] \implies \mathbf{0.32[11]}$$

**3. Adım: Hepsini Toplamak**
Bulduğumuz 4 sonucu yan yana yazıyoruz:
$$\mathbf{0.12[00] + 0.08[01] + 0.48[10] + 0.32[11]}$$

### Task 3: Vektörlerde Kronecker Çarpımı ("İçine Sokma" Mantığı)

Vektörleri tensörlediğimizde, **birinci vektörü açarız ve ikinci vektörün tamamını birinci vektörün her bir elemanının yanına yapıştırırız.**

İfademiz:
$$\begin{pmatrix} \mathbf{0.2} \\ \mathbf{0.8} \end{pmatrix} \otimes \begin{pmatrix} 0.6 \\ 0.4 \end{pmatrix}$$

**1. Adım: Birinci vektörü genişlet ve ikinciyi içine yerleştir:**
$$= \begin{pmatrix} \mathbf{0.2} \times \begin{pmatrix} 0.6 \\ 0.4 \end{pmatrix} \\ \\ \mathbf{0.8} \times \begin{pmatrix} 0.6 \\ 0.4 \end{pmatrix} \end{pmatrix}$$

**2. Adım: Skaler (dışarıdaki) sayıları parantez içine dağıt:**
$$= \begin{pmatrix} \mathbf{0.2} \times 0.6 \\ \mathbf{0.2} \times 0.4 \\ \mathbf{0.8} \times 0.6 \\ \mathbf{0.8} \times 0.4 \end{pmatrix}$$

**3. Adım: Çarpmaları yapıp 4 boyutlu tek bir sütun vektörü elde et:**
$$= \begin{pmatrix} 0.12 \\ 0.08 \\ 0.48 \\ 0.32 \end{pmatrix}$$

### Task 4: Temel Durumların (Basis States) Vektör Çarpımları

Biliyoruz ki $0$ durumu üstte $1$ taşır, $1$ durumu altta $1$ taşır:
$$[0] = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \quad \text{ve} \quad [1] = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

Bunları birleştirirken de Task 3'teki aynı "içine sokma" kuralını uyguluyoruz:

**[00] Nasıl Bulunur?**
$$[0] \otimes [0] = \begin{pmatrix} \mathbf{1} \\ \mathbf{0} \end{pmatrix} \otimes \begin{pmatrix} 1 \\ 0 \end{pmatrix} \implies \begin{pmatrix} \mathbf{1} \times \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\ \mathbf{0} \times \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{pmatrix} \implies \begin{pmatrix} \mathbf{1} \times 1 \\ \mathbf{1} \times 0 \\ \mathbf{0} \times 1 \\ \mathbf{0} \times 0 \end{pmatrix} = \mathbf{\begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}}$$
*(Gördüğün gibi 1 sayısı en üst sıraya, yani 0. indekse düştü)*

**[01] Nasıl Bulunur?**
$$[0] \otimes [1] = \begin{pmatrix} \mathbf{1} \\ \mathbf{0} \end{pmatrix} \otimes \begin{pmatrix} 0 \\ 1 \end{pmatrix} \implies \begin{pmatrix} \mathbf{1} \times \begin{pmatrix} 0 \\ 1 \end{pmatrix} \\ \mathbf{0} \times \begin{pmatrix} 0 \\ 1 \end{pmatrix} \end{pmatrix} \implies \begin{pmatrix} \mathbf{1} \times 0 \\ \mathbf{1} \times 1 \\ \mathbf{0} \times 0 \\ \mathbf{0} \times 1 \end{pmatrix} = \mathbf{\begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}}$$
*(1 sayısı ikinci sıraya düştü)*

**[10] Nasıl Bulunur?**
$$[1] \otimes [0] = \begin{pmatrix} \mathbf{0} \\ \mathbf{1} \end{pmatrix} \otimes \begin{pmatrix} 1 \\ 0 \end{pmatrix} \implies \begin{pmatrix} \mathbf{0} \times \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\ \mathbf{1} \times \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{pmatrix} \implies \begin{pmatrix} \mathbf{0} \times 1 \\ \mathbf{0} \times 0 \\ \mathbf{1} \times 1 \\ \mathbf{1} \times 0 \end{pmatrix} = \mathbf{\begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix}}$$
*(1 sayısı üçüncü sıraya düştü)*

**[11] Nasıl Bulunur?**
$$[1] \otimes [1] = \begin{pmatrix} \mathbf{0} \\ \mathbf{1} \end{pmatrix} \otimes \begin{pmatrix} 0 \\ 1 \end{pmatrix} \implies \begin{pmatrix} \mathbf{0} \times \begin{pmatrix} 0 \\ 1 \end{pmatrix} \\ \mathbf{1} \times \begin{pmatrix} 0 \\ 1 \end{pmatrix} \end{pmatrix} \implies \begin{pmatrix} \mathbf{0} \times 0 \\ \mathbf{0} \times 1 \\ \mathbf{1} \times 0 \\ \mathbf{1} \times 1 \end{pmatrix} = \mathbf{\begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}}$$
*(1 sayısı en alt sıraya düştü)*

<h3> Extra: Task 6 </h3>

<i>This task is challenging.</i>

Suppose that we have four bits.

Number 9 is represented as $ 1001 $ in binary. Verify that the vector representation of $ \pstate{1001} $ is the zero vector except its $10$th entry, which is 1.

Number 7 is represented as $ 0111 $ in binary. Verify that the vector representation of $ \pstate{0111} $ is the zero vector except its $8$th entry, which is 1.

Generalize this idea for any number between 0 and 15.

Generalize this idea for any number of bits.

**4 Bitlik Sistem ($2^4 = 16$ Durum)**

$$[1001]_2 = (1 \times 2^3) + (0 \times 2^2) + (0 \times 2^1) + (1 \times 2^0) = 8 + 0 + 0 + 1 = 9$$
Vektör gösteriminde $1$ değeri **9. indekste** yer alır. (İndeks 0'dan başladığı için baştan **10. elemandır**).

$$[0111]_2 = (0 \times 2^3) + (1 \times 2^2) + (1 \times 2^1) + (1 \times 2^0) = 0 + 4 + 2 + 1 = 7$$
Vektör gösteriminde $1$ değeri **7. indekste** yer alır. (İndeks 0'dan başladığı için baştan **8. elemandır**).

---

**Genelleme: 0 ile 15 Arasındaki Herhangi Bir Sayı İçin (4 Bit)**

4 bitlik herhangi bir durum $[b_3 b_2 b_1 b_0]$ olsun.
Ondalık karşılığı $k$ hesaplanır:
$$k = \sum_{i=0}^{3} b_i 2^i = b_3 2^3 + b_2 2^2 + b_1 2^1 + b_0 2^0$$

16 boyutlu ($16 \times 1$) durum vektörü $v$ için vektörün elemanları:
$$v_j = \begin{cases}
1 & \text{eğer } j = k \\
0 & \text{eğer } j \neq k
\end{cases} \quad \text{için } j \in \{0, 1, \dots, 15\}$$

---

**Genelleme: Herhangi Bir Bit Sayısı İçin ($n$ Bit)**

$n$ bitlik herhangi bir durum $[b_{n-1} b_{n-2} \dots b_1 b_0]$ olsun. Toplam durum sayısı $2^n$ olur.
Ondalık karşılığı $k$ hesaplanır:
$$k = \sum_{i=0}^{n-1} b_i 2^i$$

$2^n$ boyutlu ($2^n \times 1$) durum vektörü $v$ için vektörün elemanları:
$$v_j = \begin{cases}
1 & \text{eğer } j = k \\
0 & \text{eğer } j \neq k
\end{cases} \quad \text{için } j \in \{0, 1, \dots, 2^n - 1\}$$
