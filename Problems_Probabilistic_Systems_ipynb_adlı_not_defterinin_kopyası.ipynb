$$M = \begin{pmatrix} p(s_1 \to s_1) & p(s_2 \to s_1) & p(s_3 \to s_1) \\ p(s_1 \to s_2) & p(s_2 \to s_2) & p(s_3 \to s_2) \\ p(s_1 \to s_3) & p(s_2 \to s_3) & p(s_3 \to s_3) \end{pmatrix} = \mathbf{\begin{pmatrix} 0.1 & 0.45 & 0.35 \\ 0.3 & 0.4 & 0.05 \\ 0.6 & 0.15 & 0.6 \end{pmatrix}}$$

**Soru 3**
$M$ olasılıksal bir operatör ise sütun elemanlarının toplamı 1'dir.
* 1. Sütun: $0.2 + b = 1 \implies b = 0.8$
* 2. Sütun: $a + 0.3 = 1 \implies a = 0.7$

Matrisimiz: $M = \begin{pmatrix} 0.2 & 0.7 \\ 0.8 & 0.3 \end{pmatrix}$
Durum vektörümüz: $\begin{pmatrix} a \\ 1-a \end{pmatrix} = \begin{pmatrix} 0.7 \\ 1-0.7 \end{pmatrix} = \begin{pmatrix} 0.7 \\ 0.3 \end{pmatrix}$

Sonraki durumu bulmak için çarpıyoruz:
$$M \times v = \begin{pmatrix} 0.2 & 0.7 \\ 0.8 & 0.3 \end{pmatrix} \begin{pmatrix} 0.7 \\ 0.3 \end{pmatrix} = \begin{pmatrix} (0.2 \times 0.7) + (0.7 \times 0.3) \\ (0.8 \times 0.7) + (0.3 \times 0.3) \end{pmatrix} = \begin{pmatrix} 0.14 + 0.21 \\ 0.56 + 0.09 \end{pmatrix} = \mathbf{\begin{pmatrix} 0.35 \\ 0.65 \end{pmatrix}}$$
