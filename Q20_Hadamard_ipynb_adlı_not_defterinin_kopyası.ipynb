An example quantum operator for quantum coin-flipping is Hadamard.

It is defined as h-gate in Qiskit.

We implement all three experiments by using Qiskit. Here we present the first and third experiment. The second experiment will be presented later.

_This will be a warm-up step before introducing a quantum bit more formally._  

<h3> The first experiment</h3>

Our quantum bit (<b>qubit</b>) starts in state 0, which is shown as $ \ket{0} = \myvector{1 \\ 0} $.

<i>$ \ket{\cdot} $ is called ket-notation:  Ket-notation is used to represent a column vector in quantum mechanics.
For a given column vector $ \ket{v} $, its conjugate transpose is a row vector represented as $ \bra{v} $ (bra-notation).
</i>

<h4> The circuit with a single Hadamard </h4>

We design a circuit with one qubit and apply quantum coin-flipping once.

```python
!pip install -q qiskit[all] qiskit_aer pylatexenc
```

```python
# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# define a quantum register with one qubit
q =  QuantumRegister(1,"qreg")

# define a classical register with one bit
# it stores the measurement result of the quantum part
c = ClassicalRegister(1,"creg")

# define our quantum circuit
qc = QuantumCircuit(q,c)

# apply h-gate (Hadamard: quantum coin-flipping) to the first qubit
qc.h(q[0])

# measure the first qubit, and store the result in the first classical bit
qc.measure(q,c)

# draw the circuit by using matplotlib
qc.draw(output='mpl') # re-run the cell if the figure is not displayed
```

Burada 1 kuantum bitlik (qreg) ve 1 klasik bitlik (creg) en temel devreyi kuruyoruz.

qc.h(q[0]): Bu satır bütün olayın kalbidir. Kuantum bitimize Hadamard (H) kapısı uyguluyoruz. Hadamard kapısı, az önce fiziksel örneğini gördüğümüz "Işın Bölücü" (Yarı geçirgen ayna) veya "Kuantum Para Atma" işleminin koddaki tam karşılığıdır. Qubiti %50 yazı, %50 tura gelecek şekilde süperpozisyona sokar.

qc.measure(q, c): Süperpozisyondaki bu qubiti ölçerek çökertiyoruz ve sonucu (0 veya 1) klasik bite yazıyoruz.

Çıktı (Devre Çizimi): Qiskit'in çizdiği devrede qreg çizgisinin üzerinde kırmızı bir H kutusu görüyoruz. Ardından gelen siyah ölçüm cihazı simgesi, kuantum dünyasından okunan verinin alttaki çift çizgili klasik bite (creg) aktarıldığını gösteriyor.

<img src="https://github.com/sbbzplt/QWorld-Bronze/blob/master/photon/images/photon1.jpg?raw=1" width="40%">

```python
# execute the circuit 10000 times in the local simulator

job = AerSimulator().run(qc,shots=10000)
counts = job.result().get_counts(qc)
print(counts) # print the outcomes

print()
n_zeros = counts['0']
n_ones = counts['1']
print("State 0 is observed with frequency %",100*n_zeros/(n_zeros+n_ones))
print("State 1 is observed with frequency %",100*n_ones/(n_zeros+n_ones))

# we can show the result by using histogram
print()
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

The numbers of outcomes '0's and '1's are expected to be close to each other.

<h4> The circuit with two Hadamards </h4>

We design a circuit with one qubit and apply quantum coin-flipping twice.

```python
# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# define a quantum register with one qubit
q2 =  QuantumRegister(1,"qreg2")

# define a classical register with one bit
# it stores the measurement result of the quantum part
c2 = ClassicalRegister(1,"creg2")

# define our quantum circuit
qc2 = QuantumCircuit(q2,c2)

# apply h-gate (Hadamard: quantum coin-flipping) to the first qubit
qc2.h(q2[0])

# apply h-gate (Hadamard: quantum coin-flipping) to the first qubit once more
qc2.h(q2[0])

# measure the first qubit, and store the result in the first classical bit
qc2.measure(q2,c2)

# draw the circuit by using matplotlib
qc2.draw(output='mpl') # re-run the cell if the figure is not displayed
```

```python
# execute the circuit 10000 times in the local simulator

job = AerSimulator().run(qc2,shots=10000)
counts2 = job.result().get_counts(qc2)
print(counts2) # print the outcomes
```

The <u>only outcome</u> must be '0'.

Design a quantum circuit with a single qubit.

Set the value of qubit to $ \ket{1} $ by using x-gate.

Experiment 2: Apply two Hadamard gates, make measurement, and execute your program 10000 times.

The following two diagrams represent these experiments.



### Hazırlık Aşaması: Başlangıç Durumunu Ayarlama
Sistem başlangıçta $|0\rangle$ durumundadır:
$$v_0 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$

Soru bizden önce bir **X kapısı (NOT)** uygulayarak durumu $|1\rangle$ yapmamızı istiyor:
$$v_{baslangic} = X \cdot v_0 = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \mathbf{\begin{pmatrix} 0 \\ 1 \end{pmatrix}}$$
Artık her iki deney de bu $\begin{pmatrix} 0 \\ 1 \end{pmatrix}$ (yani $|1\rangle$) vektörü üzerinden başlayacak.

---

### Experiment 1: Tek Hadamard Kapısı (10.000 Atış)
Bu adımda sisteme sadece bir kez Hadamard ($H$) uyguluyoruz ve ölçüm yapıyoruz.

$$v_{exp1} = H \cdot v_{baslangic} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

Çarpmayı yapalım:
$$v_{exp1} = \begin{pmatrix} (\frac{1}{\sqrt{2}} \cdot 0) + (\frac{1}{\sqrt{2}} \cdot 1) \\ (\frac{1}{\sqrt{2}} \cdot 0) + (-\frac{1}{\sqrt{2}} \cdot 1) \end{pmatrix} = \mathbf{\begin{pmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{pmatrix}}$$

**Ölçüm Olasılıkları (Genliklerin Karesi):**
* $|0\rangle$ gelme ihtimali: $\left(\frac{1}{\sqrt{2}}\right)^2 = \mathbf{0.5 \ (\%50)}$
* $|1\rangle$ gelme ihtimali: $\left(-\frac{1}{\sqrt{2}}\right)^2 = \mathbf{0.5 \ (\%50)}$

**Sonuç Tahmini:** Simülatörü 10.000 kez çalıştırdığında yaklaşık **~5000 kez 0**, **~5000 kez 1** ölçeceksin.

---

### Experiment 2: İki Hadamard Kapısı Üst Üste (10.000 Atış)
Bu adımda, az önce birinci Hadamard'dan çıkan o süperpozisyon durumundaki vektöre, ölçüm yapmadan *hemen önce* ikinci bir Hadamard kapısı daha uyguluyoruz.

$$v_{exp2} = H \cdot v_{exp1} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{pmatrix}$$

Çarpmayı yapalım ve kuantum girişimini (interference) görelim:
* **Üst Satır ($|0\rangle$ ihtimali):** $\left(\frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}\right) + \left(\frac{1}{\sqrt{2}} \cdot -\frac{1}{\sqrt{2}}\right) = \frac{1}{2} - \frac{1}{2} = \mathbf{0}$
* **Alt Satır ($|1\rangle$ ihtimali):** $\left(\frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}\right) + \left(-\frac{1}{\sqrt{2}} \cdot -\frac{1}{\sqrt{2}}\right) = \frac{1}{2} + \frac{1}{2} = \mathbf{1}$

$$v_{exp2} = \mathbf{\begin{pmatrix} 0 \\ 1 \end{pmatrix}}$$

**Ölçüm Olasılıkları:**
* $|0\rangle$ gelme ihtimali: $0^2 = \mathbf{0 \ (\%0)}$
* $|1\rangle$ gelme ihtimali: $1^2 = \mathbf{1.0 \ (\%100)}$

**Sonuç Tahmini:** Simülatörü 10.000 kez çalıştırdığında **0 kez 0**, tam **10.000 kez 1** ölçeceksin.

---

```python
#EXP111


from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# 1 qubitlik ve 1 klasik bitlik devre oluştur
q = QuantumRegister(1, "q")
c = ClassicalRegister(1, "c")
qc1 = QuantumCircuit(q, c)

# Başlangıç ayarı: X kapısı ile qubiti |1> durumuna getir
qc1.x(q[0])

# DENEY 1: Sadece BİR tane Hadamard kapısı uygula
qc1.h(q[0])

# Ölçüm yap ve sonucu klasik bite yaz
qc1.measure(q, c)

# Devreyi lokal simülatörde 10.000 kez çalıştır
job1 = AerSimulator().run(qc1, shots=10000)
counts1 = job1.result().get_counts(qc1)

# Sonucu yazdır
print("Deney 1 Sonuçları (Tek Hadamard):", counts1)
```

```python
###EXP2222

# Yeni bir devre oluştur (aynı q ve c hafızalarını kullanabiliriz)
qc2 = QuantumCircuit(q, c)

# Başlangıç ayarı: X kapısı ile qubiti |1> durumuna getir
qc2.x(q[0])

# DENEY 2: İKİ tane Hadamard kapısını peş peşe uygula
qc2.h(q[0]) # İlk Işın Bölücü (Süperpozisyon)
qc2.h(q[0]) # İkinci Işın Bölücü (Girişim / Eski haline dönme)

# Ölçüm yap
qc2.measure(q, c)

# Devreyi simülatörde 10.000 kez çalıştır
job2 = AerSimulator().run(qc2, shots=10000)
counts2 = job2.result().get_counts(qc2)

# Sonucu yazdır
print("Deney 2 Sonuçları (İki Hadamard):", counts2)
```
