_Quantum systems are linear systems: "The quantum states are represented by vectors and quantum operators are represented by matrices. The new quantum states are calculated by corresponding matrix-vector multiplications."_

A qubit (quantum bit) has two states: state 0 and state 1.

They are denoted by ket-notation:

$ \ket{0} = \myvector{1 \\ 0} $ and $ \ket{1} = \myvector{0\\ 1} $.



<h3> NOT operator </h3>

NOT operator flips the value of a qubit.

We use capital letter for the matrix form of the operators:

$ X = \X$.

The action of $ X $ on the qubit:

$ X \ket{0} = \ket{1} $.

More explicitly, $ X \ket{0} = \X \vzero = \vone = \ket{1} $.

Similarly, $ X \ket{1} = \ket{0} $.

More explicitly, $ X \ket{1} = \X \vone = \vzero = \ket{0} $.

<b><i>Why is the NOT operator referred to as the x-gate?<i></b>
   
In Bronze, we use only real numbers, but we should note that complex numbers are also used in quantum computing. When complex numbers are used, the state of a qubit can be represented by a four dimensional real number valued vector, which is not possible to visualize. On the other hand, it is possible to represent such state in three dimensions (with real numbers). This representation is called [Bloch sphere](https://en.wikipedia.org/wiki/Bloch_sphere).

In three dimensions, we have axes: x, y, and z. X refers to the rotation with respect to x-axis. Similarly, we have the rotation with respect to y-axis and z-axis. In Bronze, we will also see the operator Z (z-gate). The operator Y is defined with complex numbers.

Görseldeki notlar, kuantum mekaniğinin en temel yapı taşları olan **Qubit (Kuantum Biti)** gösterimlerini ve en temel kapı olan **X Kapısı'nın (NOT Operatörü)** matematiğini anlatıyor.

Hiçbir adımı atlamadan, o matris çarpmalarının (ara işlemlerin) nasıl yapıldığını ve notun sonundaki o ilginç "Neden X kapısı diyoruz?" sorusunun cevabını tane tane açalım.

### 1. Qubitlerin Vektör Gösterimi (Ket Notasyonu)
Kuantumda durumları göstermek için matematikteki $| \ \rangle$ (Ket) sembolünü kullanırız.
* **State 0 (Durum 0):** $|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$
* **State 1 (Durum 1):** $|1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$

---

### 2. NOT Operatörü (X Kapısı)
Klasik bilgisayarlardaki 0'ı 1, 1'i 0 yapan NOT kapısının kuantum matris formudur:
$$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$

#### İşlem 1: X Kapısını $|0\rangle$ Durumuna Uygulamak
Amacımız $X|0\rangle$ işlemini yapmak. Matris çarpımı kuralı: Satır ile Sütun karşılıklı çarpılır ve toplanır.

$$X|0\rangle = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$

**Ara İşlemler (Satır $\times$ Sütun):**
* **Üst Satır:** $(0 \times 1) + (1 \times 0) = 0 + 0 = \mathbf{0}$
* **Alt Satır:** $(1 \times 1) + (0 \times 0) = 1 + 0 = \mathbf{1}$

**Sonuç:**
$$\begin{pmatrix} 0 \\ 1 \end{pmatrix} = \mathbf{|1\rangle}$$
*(Gördüğün gibi 0 durumunu aldı, 1 durumuna çevirdi).*

#### İşlem 2: X Kapısını $|1\rangle$ Durumuna Uygulamak
Şimdi aynı matrisi $X|1\rangle$ işlemi için kullanalım.

$$X|1\rangle = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

**Ara İşlemler (Satır $\times$ Sütun):**
* **Üst Satır:** $(0 \times 0) + (1 \times 1) = 0 + 1 = \mathbf{1}$
* **Alt Satır:** $(1 \times 0) + (0 \times 1) = 0 + 0 = \mathbf{0}$

**Sonuç:**
$$\begin{pmatrix} 1 \\ 0 \end{pmatrix} = \mathbf{|0\rangle}$$
*(1 durumunu aldı, 0 durumuna çevirdi).*

---

### 3. Neden Bu Kapının Adı "X" Kapısı? (Görselin Alt Kısmı)
Klasik mantıkta buna sadece "NOT" (Değil) kapısı deyip geçeriz. Ama kuantum dünyasında buna **X-gate** denmesinin geometrik bir sebebi var.

Notta açıklanan şey özetle şudur:
Kuantum bitleri sadece 0 ve 1'den ibaret değildir; aradaki sonsuz süperpozisyon durumlarını da barındırırlar. Bu karmaşık durumları gözümüzde canlandırabilmek için **Bloch Küresi (Bloch Sphere)** adını verdiğimiz 3 boyutlu bir küre kullanırız.

Bu kürenin bildiğimiz 3 ekseni vardır: **x, y ve z eksenleri.**
* Kürenin en tepesi $|0\rangle$ durumudur (Kuzey Kutbu).
* Kürenin en altı $|1\rangle$ durumudur (Güney Kutbu).

Sen bu kürenin en tepesindeki $|0\rangle$ noktasını en altındaki $|1\rangle$ noktasına getirmek istersen, o küreyi **x-ekseni etrafında 180 derece (veya pi radyan) döndürmen** gerekir.

İşte bu yüzden, qubitin değerini tersine çeviren bu matrise, Bloch Küresi'nde "X ekseni etrafında döndürme" işlemi yaptığı için **X Operatörü** adı verilir. Z ekseni etrafında döndürenlere Z-gate, Y ekseni etrafında döndürenlere de Y-gate diyeceğiz.

<h3> Hadamard operator</h3>

Hadamard operator ($ H $ or h-gate) looks similar to a fair coin-flipping.

$$
    H = \hadamard.
$$

But, there are certain dissimilarities:
<ul>
    <li> we have a <u>negative entry</u>, and</li>
    <li> instead of $ \frac{1}{2} $, we have <u>its square root</u> $ \mypar{ \frac{1}{\sqrt{2}} } $. </li>
</ul>

<font color="blue"> Quantum systems can have negative transitions. </font>

<font color="blue"> Can probabilistic systems be extended with negative values?</font>

1. Hadamard Operatörü ve Klasik Paradan FarkıNotlarda belirtildiği gibi, Hadamard kapısı hilesiz bir para atmaya (fair coin-flipping) benzer ama iki büyük farkı vardır:Matrisin içinde eksi (negatif) değerli bir giriş vardır ($-\frac{1}{\sqrt{2}}$).Olasılık değeri olan $\frac{1}{2}$ (yani %50) yerine, onun karekökü olan $\frac{1}{\sqrt{2}}$ değerini kullanır.Bunun sebebi kuantum mekaniğinde olasılıklarla değil, dalga formundaki "olasılık genlikleriyle" (probability amplitudes) işlem yapmamızdır.

<h4> One-step Hadamard</h4>

Start in $ \ket{0} $.

After applying $ H $:

$$
  H \ket{0} =  \hadamard \vzero =  \stateplus.
$$
<font color="blue"> Mini Task: </font> Write above result in terms of unit states, _i.e._ $\ket0$ and $\ket1$.

After measurement, we observe the states $ \ket{0} $ and $ \ket{1} $ with equal probability $ \frac{1}{2} $.

How can this be possible when their values are $ \frac{1}{\sqrt{2}} $?


Vektörün üst elemanı $|0\rangle$ durumunun katsayısını, alt elemanı $|1\rangle$ durumunun katsayısını verir.
O halde matematiksel formdaki bu vektörü Dirac (ket) notasyonu ile şu şekilde yazarız:

Cevap: $\mathbf{\frac{1}{\sqrt{2}}|0\rangle + \frac{1}{\sqrt{2}}|1\rangle}$

2. Tek Adımlı Hadamard (Ara İşlemlerle Çarpım)Sistemimiz $|0\rangle$ durumunda başlıyor:$$|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$Bu duruma Hadamard ($H$) operatörünü uyguluyoruz ($H|0\rangle$):$$\begin{pmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$Çarpımı adım adım açalım (Satır $\times$ Sütun):

Üst Satır: $\left(\frac{1}{\sqrt{2}} \times 1\right) + \left(\frac{1}{\sqrt{2}} \times 0\right) = \frac{1}{\sqrt{2}} + 0 = \mathbf{\frac{1}{\sqrt{2}}}$

Alt Satır: $\left(\frac{1}{\sqrt{2}} \times 1\right) + \left(-\frac{1}{\sqrt{2}} \times 0\right) = \frac{1}{\sqrt{2}} + 0 = \mathbf{\frac{1}{\sqrt{2}}}$

Sonuç Vektörü:$$\mathbf{\begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{pmatrix}}$$

Let's start in $ \ket{1} $.

After applying $ H $:

$$
    H \ket{1} = \hadamard \vone = \stateminus.
$$

<font color="blue"> Mini Task: </font> Write above result in terms of unit states, _i.e._ $\ket0$ and $\ket1$.

After measurement, we observe the states $ \ket{0} $ and $ \ket{1} $ with equal probability $ \frac{1}{2} $.

We obtain the same values even when one of the values is negative.

<img src="https://github.com/sbbzplt/QWorld-Bronze/blob/master/photon/images/photon3c.jpg?raw=1" width="35%" align="left">

<i>The absolute value of a negative value is positive.</i>

<i>The square of a negative value is also positive.</i>

As we have observed, the second fact fits better when reading the measurement results.

<font color="blue"><b> When a quantum system is measured, the probability of observing one state is the square of its value [*].</b></font>

The value of the system being in a state is called its <b>amplitude</b>.

In the above example, the amplitudes of states $\ket{0}$ and $ \ket{1} $ are respectively $ \sqrttwo $ and $ -\sqrttwo $.

The probabilities of observing them after a measurement are $ \onehalf $.

_Remark that, after observing state $ 0 $, the new state will be $ \ket{0} $, and, after observing state $ 1 $, the new state will be $ \ket{1} $._

_[*] the square of its magnitude value when using complex numbers_

### States $ \ket{+} $ and $ \ket{-} $

In the literature, the quantum states obtained after applying $ H $ to $ \ket{0} $ and $ \ket{1} $ are known as **ket-plus** ($\ket{+}$) and **ket-minus** ($ \ket{-} $) states, respectively:

$ \ket{+} = H \ket{0} = \hadamard \vzero = \stateplus $

$ \ket{-} = H \ket{1} =  \hadamard \vone = \stateminus $
