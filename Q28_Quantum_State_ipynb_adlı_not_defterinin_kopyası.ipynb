Şimdi elinde iki durumlu (0 ve 1) bir kuantum sistemi var ve vektörün $\begin{pmatrix} a \\ b \end{pmatrix}$ şeklinde.Born Kuralı'na göre:$|0\rangle$ gelme olasılığı: $a^2$$|1\rangle$ gelme olasılığı: $b^2$Peki fizikte değişmez bir yasa nedir? Tüm ihtimallerin toplamı her zaman %100, yani 1 olmalıdır. Sistem ya 0'dır ya da 1'dir, kaybolamaz.Bu durumda denklemimiz kendiliğinden ortaya çıkar:$$a^2 + b^2 = 1$$Bu denklem sana tanıdık geldi mi?Bu, dik bir üçgenin hipotenüsünü veya bir çemberin yarıçapını bulduğumuz Pisagor Teoremidir ($x^2 + y^2 = r^2$).


Born Kuralı der ki: Bir kuantum durumunu ölçtüğünüzde, belirli bir durumu bulma olasılığınız, o duruma ait genliğin mutlak karesine eşittir.Matematiksel olarak:$$\text{Olasılık} = |\text{Genlik}|^2$$

**Bra-ket notasyonu**, kuantum mekaniğinin evrensel dilidir. Ünlü fizikçi Paul Dirac tarafından bulunduğu için **Dirac Notasyonu** olarak da bilinir.

Kuantum hesaplamalarında sürekli matrisler ve uzun vektörler çizmek çok yorucu ve sayfalar dolusu yer kaplayan bir iş olduğu için, bilim insanları bu işi kısaltmak amacıyla bu sembolik dili geliştirmiştir.

İsmi, İngilizce parantez anlamına gelen **"bracket"** $(\langle \ \ \rangle)$ kelimesinin ortadan ikiye bölünmesinden gelir: **Bra** $\langle \ |$ ve **Ket** $| \ \rangle$.

Bu iki parçayı ve birleştiklerinde ne anlama geldiklerini tane tane inceleyelim:

### 1. Ket: $| \ \rangle$
Ket, bildiğimiz bir **sütun vektörüdür**. Kuantum sisteminin o anki fiziksel durumunu (state) temsil eder. İçinde o durumun genliklerini barındırır.
Örneğin, 0 durumu bir "ket" olarak şöyle yazılır:
$$|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$
Veya genel bir $u$ durumu:
$$|u\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$$

### 2. Bra: $\langle \ |$
Bra, ket'in yere yatırılmış hali, yani bir **satır vektörüdür**. Matematiksel olarak, ket vektörünün devriği (transpose) alınarak oluşturulur.
Örneğin, 0 durumunun "bra" hali şöyle yazılır:
$$\langle 0| = \begin{pmatrix} 1 & 0 \end{pmatrix}$$
Veya $u$ durumunun bra hali:
$$\langle u| = \begin{pmatrix} a & b \end{pmatrix}$$

---

### 3. Bra-Ket (Kavuşma Anı): $\langle \ | \ \rangle$
Bir "bra" (satır) ile bir "ket" (sütun) yan yana geldiğinde, matematikteki **İç Çarpım (Inner Product)** işlemi gerçekleşir. Satır ile sütun çarpılır ve sonuç her zaman **tek bir skaler sayı** çıkar.

Az önce seninle konuştuğumuz o "Normalizasyon / Pisagor" kuralını ($a^2 + b^2 = 1$) hatırla. Notlarda yer alan $\langle u | u \rangle = 1$ ifadesinin bra-ket notasyonuyla nasıl açıldığını görelim:

$|u\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$ ve $\langle u| = \begin{pmatrix} a & b \end{pmatrix}$ ise;

$$\langle u | u \rangle = \begin{pmatrix} a & b \end{pmatrix} \begin{pmatrix} a \\ b \end{pmatrix}$$

Çarpma işlemini (satır $\times$ sütun) yapalım:
$$\langle u | u \rangle = (a \cdot a) + (b \cdot b)$$
$$\mathbf{\langle u | u \rangle = a^2 + b^2 = 1}$$

Gördüğün gibi, sayfalar dolusu matris çizmek veya "bir vektörün kendisiyle iç çarpımı 1'e eşittir" diye uzun uzun cümleler kurmak yerine, fizikçiler tahtaya sadece **$\langle u | u \rangle = 1$** yazarlar. Bra-ket notasyonu, bu karmaşık matematiği bir sembole sıkıştıran son derece zarif bir araçtır.

_The overall probability must be 1 when we observe a quantum system._

For example, the following vectors <u>cannot</u> be a valid quantum state:

$$
    \myvector{ \dfrac{1}{2} \\ \dfrac{1}{2} }
    \mbox{ and }
    \myvector{ \dfrac{\sqrt{3}}{2} \\ \dfrac{1}{\sqrt{2}} }.
$$

For the first vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are $ \dfrac{1}{4} $.

So, the overall probability of getting a result is $ \dfrac{1}{4} + \dfrac{1}{4} = \dfrac{1}{2} $, which is less than 1.

For the second vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are respectively $ \dfrac{3}{4} $ and $ \dfrac{1}{2} $.

So, the overall probability of getting a result is $ \dfrac{3}{4} + \dfrac{1}{2} = \dfrac{5}{4} $, which is greater than 1.

<font color="blue"><b>The summation of amplitude squares must be 1 for a valid quantum state.</b></font>

<font color="blue"><b>More formally, a quantum state can be represented by a vector having length 1, and vice versa.</b></font>

<i>The summation of amplitude squares gives the square of the length of vector.

But, this summation is 1, and its square root is also 1. So, we can use the term <u>length</u> in the definition.</i>

<i> <b>Technical notes:</b> We represent a quantum state as $ \ket{u} $ instead of $ u $. Remember the relation between the length and dot product: $ \norm{u} = \sqrt{\dot{u}{u}} $.
    
In quantum computation, we use inner product instead of dot product, which is defined on complex numbers. By using bra-ket notation, $ \norm{ \ket{u} } = \sqrt{ \braket{u}{u} } = 1 $, or equivalently $ \braket{u}{u} = 1 $, where $ \braket{u}{u} $ is a short form of $ \bra{u}\ket{u} $. For real-valued vectors, $ \braket{v}{v} = \dot{v}{v}  $.
</i>

<h3> Task 1 </h3>

If the following vectors are valid quantum states defined with real numbers, then what can be the values of $a$ and $b$?

$$
    \ket{v} = \myrvector{a \\ -0.1 \\ -0.3 \\ 0.4 \\ 0.5}
    ~~~~~ \mbox{and} ~~~~~
   \ket{u} = \myrvector{ \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{b}} \\ -\frac{1}{\sqrt{3}} }.
$$

Bir kuantum durumunun geçerli olabilmesi için içindeki tüm elemanların (genliklerin) karelerinin toplamı kesinlikle 1 olmalıdır.1. Vektör $|v\rangle$ için $a$'nın BulunmasıVektörümüz 5 elemanlı: $a$, $-0.1$, $-0.3$, $0.4$, $0.5$.Karelerini alıp 1'e eşitliyoruz:$$a^2 + (-0.1)^2 + (-0.3)^2 + 0.4^2 + 0.5^2 = 1$$Kare alma işlemlerini yapalım:$$a^2 + 0.01 + 0.09 + 0.16 + 0.25 = 1$$Sayıları toplayalım:$$a^2 + 0.51 = 1$$$a^2$'yi yalnız bırakalım:$$a^2 = 1 - 0.51$$$$a^2 = 0.49$$Kök dışına çıkarırken her zaman hem pozitif hem negatif durumu alırız (kuantumda genlikler eksi olabilir demiştik):$$\mathbf{a = 0.7} \quad \text{veya} \quad \mathbf{a = -0.7}$$2. Vektör $|u\rangle$ için $b$'nin BulunmasıVektörümüz 3 elemanlı (kesirli sayılar olduğu için parantez daha uzun çizilmiş): $\frac{1}{\sqrt{2}}$, $\frac{1}{\sqrt{b}}$, $-\frac{1}{\sqrt{3}}$.Aynı kuralı uygulayıp karelerini 1'e eşitliyoruz:$$\left(\frac{1}{\sqrt{2}}\right)^2 + \left(\frac{1}{\sqrt{b}}\right)^2 + \left(-\frac{1}{\sqrt{3}}\right)^2 = 1$$Karelerini aldığımızda kökler ortadan kalkar (ve eksi işaret artıya döner):$$\frac{1}{2} + \frac{1}{b} + \frac{1}{3} = 1$$Bilinen kesirleri ($\frac{1}{2}$ ve $\frac{1}{3}$) toplayalım. Paydaları 6'da eşitlersek ($\frac{3}{6} + \frac{2}{6}$):$$\frac{5}{6} + \frac{1}{b} = 1$$$\frac{1}{b}$'yi yalnız bırakalım:$$\frac{1}{b} = 1 - \frac{5}{6}$$$$\frac{1}{b} = \frac{1}{6}$$Buradan doğrudan sonuç çıkar:$$\mathbf{b = 6}$$

```python
# Vektör |v> için a'nın hesaplanması
bilinenlerin_kare_toplami = (-0.1)**2 + (-0.3)**2 + (0.4)**2 + (0.5)**2
a_kare = 1 - bilinenlerin_kare_toplami

# Kök dışına çıkarma (hem artı hem eksi)
a_pozitif = a_kare ** 0.5
a_negatif = -a_pozitif

print("a'nın alabileceği değerler:", round(a_pozitif, 2), "ve", round(a_negatif, 2))

# Vektör |u> için b'nin hesaplanması
bilinen_kesirler_toplami = (1/(2**0.5))**2 + (-1/(3**0.5))**2
b_tersi = 1 - bilinen_kesirler_toplami

b = 1 / b_tersi

print("b'nin değeri:", round(b))
```

```python
import random

# 1. Görev: Rastgele geçerli bir kuantum durumu (vektör) üreten fonksiyon
def random_quantum_state():
    # -100 ile 100 arasında iki rastgele sayı seç
    x = random.uniform(-100, 100)
    y = random.uniform(-100, 100)

    # 0,0 gelirse tekrar seç (Sıfıra bölme hatasını önlemek için)
    while x == 0 and y == 0:
        x = random.uniform(-100, 100)
        y = random.uniform(-100, 100)

    # Uzunluğu (normalization factor) bul (Pisagor)
    length = (x**2 + y**2)**0.5

    # Elemanları uzunluğa bölerek normalize et (Kareleri toplamı artık 1 olacak)
    return [x/length, y/length]

# 2. Görev: Verilen vektörün geçerli bir kuantum durumu olup olmadığını test eden fonksiyon
def is_quantum_state(state):
    # Genliklerin karelerini topla
    sum_of_squares = state[0]**2 + state[1]**2

    # Bilgisayardaki küsurat hassasiyetinden (precision) dolayı tam 1'e değil, 1'e çok yakın mı diye bakıyoruz
    if abs(sum_of_squares - 1.0) < 0.00001:
        return True
    else:
        return False

# Hadamard Operatörünü uygulayan yardımcı fonksiyon
def apply_hadamard(state):
    h_factor = 1 / (2**0.5)
    # Satır x Sütun çarpımı: H * State
    new_state_0 = (h_factor * state[0]) + (h_factor * state[1])
    new_state_1 = (h_factor * state[0]) - (h_factor * state[1])
    return [new_state_0, new_state_1]

# 3. Görev: Döngüyü 10 kere tekrarla
for i in range(10):
    print(f"--- Deneme {i+1} ---")

    # Rastgele bir kuantum durumu seç
    my_state = random_quantum_state()
    print(f"Üretilen Vektör: [{my_state[0]:.4f}, {my_state[1]:.4f}]")

    # İlk durumu kontrol et
    is_valid_before = is_quantum_state(my_state)
    print(f"Başlangıçta Geçerli mi (Uzunluk 1 mi)?: {is_valid_before}")

    # Rastgele duruma Hadamard matrisini uygula
    new_state = apply_hadamard(my_state)
    print(f"Hadamard Sonrası Vektör: [{new_state[0]:.4f}, {new_state[1]:.4f}]")

    # Hadamard'dan sonra durumu tekrar kontrol et
    is_valid_after = is_quantum_state(new_state)
    print(f"Hadamard Sonrası Geçerli mi (Uzunluk Korundu mu)?: {is_valid_after}\n")
```
