```python
import matplotlib.pyplot as plt
import random

# Önceki notebook'tan hatırladığın rastgele kuantum durumu üreten fonksiyon
def random_quantum_state():
    x = random.uniform(-100, 100)
    y = random.uniform(-100, 100)
    # Normalizasyon faktörü (uzunluk)
    length = (x**2 + y**2)**0.5
    return [x/length, y/length]

# Çizim alanı hazırlığı
plt.figure(figsize=(6,6), dpi=80)
plt.axhline(0, color='black', linewidth=1) # x ekseni
plt.axvline(0, color='black', linewidth=1) # y ekseni

# Renk listesi
colors = ['r', 'b', 'g', 'y', 'c', 'm']

# 1. Aşama: 100 adet kuantum durumu çiz
for i in range(100):
    state = random_quantum_state()
    plt.plot(state[0], state[1], marker='o', color=random.choice(colors), markersize=4)

# 2. Aşama: 1000 adet kuantum durumu çiz
for i in range(1000):
    state = random_quantum_state()
    # Şeffaflık (alpha) ekleyelim ki üst üste binenler çemberi netleştirsin
    plt.plot(state[0], state[1], marker='.', color=random.choice(colors), markersize=2, alpha=0.5)

plt.title("Rastgele Kuantum Durumlarının Görselleştirilmesi")
plt.xlabel("State 0 Genliği (x)")
plt.ylabel("State 1 Genliği (y)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

<h3> The angle of a quantum state </h3>

The angle of a vector (in radians) on the unit circle is the length of arc in counter-clockwise direction that starts from $ (1,0) $ and with the points representing the vector.

We execute the following code a couple of times to see different examples, where the angle is picked randomly in each run.

You can also set the value of "myangle" manually for seeing a specific angle.

1. Mantık (Konsept)
Bu görev qubit'in Bloch çemberi (2D birim çember) üzerinde görselleştirilmesiyle ilgili.
Bir qubit'in durumu şöyle ifade edilir:

|ψ⟩ = cos(θ)|0⟩ + sin(θ)|1⟩

Burada θ (theta) bir açıdır. Bu açıyı rastgele seçerek qubit durumu oluşturuyoruz ve bunu birim çember üzerinde bir nokta/ok olarak çiziyoruz.

2. Matematik (Ara İfadeler Dahil)
Adım 1 — Rastgele açı seç:
θ ∈ {0°, 1°, 2°, ..., 360°}  →  randrange(0, 360) ile seçilir
Adım 2 — Radyana çevir:
θ_rad = θ × (π / 180)
Adım 3 — Qubit amplitüdlerini hesapla:
α = cos(θ_rad)   →  |0⟩ bileşeni (x ekseni)
β = sin(θ_rad)   →  |1⟩ bileşeni (y ekseni)
Adım 4 — Normalizasyon kontrolü:
|α|² + |β|² = cos²(θ) + sin²(θ) = 1  ✓
Adım 5 — Kartezyen koordinat:
x = cos(θ_rad)
y = sin(θ_rad)
nokta: (x, y)  →  birim çember üzerinde

```python
# %%writefile random_qstate.py   ← kaydetmek istersen bunu aç
from math import cos, sin, pi
from random import randrange

def random_qstate_by_angle():
    """
    Rastgele bir açı seçerek qubit durumu oluşturur.
    Döndürür: (x, y) → birim çember üzerindeki koordinatlar
    """
    # Adım 1: Rastgele derece cinsinden açı (0–359)
    degree = randrange(0, 360)

    # Adım 2: Radyana çevir
    theta = degree * pi / 180

    # Adım 3: Amplitüdleri hesapla
    x = cos(theta)   # α  — |0⟩ bileşeni
    y = sin(theta)   # β  — |1⟩ bileşeni

    return x, y
```
